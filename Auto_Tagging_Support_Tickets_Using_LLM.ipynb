```python
import pandas as pd
from transformers import pipeline
```

```python
data = {
    "ticket": [
        "My internet connection is very slow",
        "I forgot my account password",
        "The payment page is not loading",
        "My order has not arrived yet",
        "The mobile app keeps crashing"
    ]
}

df = pd.DataFrame(data)

df.head()
```

```python
labels = [
    "technical issue",
    "billing problem",
    "account issue",
    "delivery issue",
    "general inquiry"
]
```

```python
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)
```

```python
results = []

for ticket in df["ticket"]:

    prediction = classifier(ticket, labels)

    top3 = prediction["labels"][:3]

    results.append(top3)

df["top_3_tags"] = results

df
```

```python
for i in range(len(df)):
    print("Ticket:", df["ticket"][i])
    print("Predicted Tags:", df["top_3_tags"][i])
    print()
```

```python
df.to_csv("tagged_tickets.csv", index=False)
```
